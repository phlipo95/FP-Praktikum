```python
%matplotlib inline
from scipy.optimize import curve_fit
from uncertainties import ufloat
from uncertainties.unumpy import uarray
from uncertainties import unumpy
import scipy.constants as con
```

```python
MHz_10 = np.array([[0,100, 200, 300, 400, 460],[5911,4572,3351,2070,792,52]])
MHz_15 = np.array([[201, 300, 403, 500, 600, 670], [5980, 4763, 3477, 2245, 960, 120]])
MHz_20 = np.array([[200, 303, 401, 499, 600, 700, 801],[5928, 5018, 4064, 3138, 2245, 1246, 362]])
MHz_25 = np.array([[300, 404, 496, 602, 700, 805],[5823, 4830, 3815, 2870, 1953, 896]])
MHz_30 = np.array([[300, 400, 496, 601, 700, 800, 900],[6062, 5220, 4308, 3470, 2528 ,1624 ,762]])
```

```python
def f(x, m, b):
    return m*x + b

def magnFluss(I):
    return 8/np.sqrt(125)*con.mu_0*156/0.1*I

def calcGyro(B, nu):
    return con.h * nu / (9.274*10**(-24) * B)

def scaleing(pixel, strom, nu, mes1=0, mes2=0, savefig='test'):
    params, cov = curve_fit(f, pixel, strom)
    
    xlims = np.array([min(pixel) - 0.05*(max(pixel)), max(pixel) + 0.05*(max(pixel))])
    ylims = np.array([min(strom) - 0.05*(max(strom)), max(strom) + 0.05*(max(strom))])
    plt.plot(pixel, strom, 'gx', label=r'Messwerte Skalierung')
    plt.plot(np.linspace(*xlims, 30), f(np.linspace(*xlims, 30), *params), 'b--', label=r'lineare Regression')
    plt.plot(mes1, f(mes1, *params), 'ro', label='entgegen magn. Feld der Erde', alpha= 0.6)
    plt.plot(mes2, f(mes2, *params), 'yo', label='mit dem magn. Feld der Erde', alpha = 0.6)
    plt.xlim(*xlims)
    plt.ylim(*ylims)
    plt.xlabel(r'Pixel vom eingescannten Bild')
    plt.ylabel(r'Strom in mA')
    plt.grid()
    plt.legend(loc='best')
    
    gradient = ufloat(params[0], np.sqrt(np.diag(cov)[0]))
    bios = ufloat(params[1], np.sqrt(np.diag(cov)[1]))
    
    print('Gradient: ', gradient, ', bios: ', bios)
    strom1 = f(mes1, gradient, bios)/1000
    print('Strom 1: ', strom1)
    strom2 = f(mes2, gradient, bios)/1000
    print('Strom 2: ', strom2)
    bFeld =  (magnFluss(strom1 - strom2))/2
    print('B-Feld der Erde: ', bFeld)
    print('B-Feldst채rke mit dem Magnf. :', magnFluss(strom1))
    print('B-Feldst채rke entgegen das Magnf. :', magnFluss(strom2))
    meanBFeld = magnFluss(strom1) - bFeld
    print('Gemitteltes B-Feld: ', meanBFeld)
    print('gyromagnetischer Faktor: ', calcGyro(meanBFeld, nu))
    
    return bFeld, nu, calcGyro(meanBFeld, nu), meanBFeld
```

```python
magn10, f10, g10, b10= scaleing(MHz_10[1], MHz_10[0], 10.62*10**6, 2101, 2630)
```

```python
magn15, f15, g15, b15 = scaleing(MHz_15[1], MHz_15[0], 15.94*10**6, 3234, 3887)
```

```python
magn20, f20, g20, b20 = scaleing(MHz_20[1], MHz_20[0], 20.54*10**6, 2937, 3404)
```

```python
magn25, f25, g25, b25 = scaleing(MHz_25[1], MHz_25[0], 25.02*10**6, 2198, 2900)
```

```python
magn30, f30, g30, b30 = scaleing(MHz_30[1], MHz_30[0], 30*10**6, 1756, 2254)
```

```python
np.mean([g10,g15,g20,g25,g30])
```

```python
np.mean([magn10,magn15,magn20,magn25,magn30])
```

```python
frequ = np.array([f10,f15,f20,f25,f30])
bFeld = np.array([b10,b15,b20,b25,b30])

params, cov = curve_fit(f, unumpy.nominal_values(frequ), unumpy.nominal_values(bFeld))

#plt.plot(unumpy.nominal_values(frequ), unumpy.nominal_values(bFeld), 'x', label=r'Quotient')
plt.errorbar(unumpy.nominal_values(frequ),
             unumpy.nominal_values(bFeld),
             xerr=unumpy.std_devs(frequ), 
             yerr=unumpy.std_devs(bFeld), 
             fmt='.',
             label='Messwerte mit Fehler')
plt.plot(unumpy.nominal_values(frequ), f(unumpy.nominal_values(frequ), *params), label=r'Fit')
plt.legend(loc='best')
plt.xlabel(r'Frequenz $\nu_e$ / $\frac{1}{s}$')
plt.ylabel(r'gemitteltes B-Feld / T')
plt.savefig('picture/lande.pdf')
```
